```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# handling missing values

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month':0, 'last review': df['last review'].min()}, inplace = True)
```

```python
df.dropna(subset = ['NAME', 'host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=["license", "house_rules"], errors='ignore')
```

```python
df.head()
```

# remove dollar signs and convert to float

```python
df['price'] = df['price'].replace('[\$,]','',regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]','',regex=True).astype(float)
```

```python
df.head()
```

# remove duplicates

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

# decriptive statistics

```python
df.describe()
```

# visualisation

# what is the distribution of listing prices?

```python
plt.figure(figsize = (10,6))
sns.histplot(df['price'], bins= 50,kde = True,color = 'red')
plt.title("Distribution of listing price")
plt.xlabel('price $')
plt.ylabel('frequency')
plt.show()
```

```python
The histogram shows a fairly even distribution of listing prices across different price ranges,
indicating no particular concentration of listing in any specific range.
The KDE line helps visualize this even spread more clearly,
confirming that the dataset contains listings with a wide variety of prices.
    
```

# How are different room types distributed?

```python
df['room type']
```

```python
plt.figure(figsize = (8,5))
sns.countplot ( x= 'room type', data = df, color = 'pink')
plt.title('room type distribution')
plt.xlabel('room type')
plt.ylabel('count')
plt.show()
```

```python
Frome above it is clear that the entire home or apartment are most preferred
followed by private room then shared room and leastly the hotel room.
```

# How are the listings distributed across different neighborhoods?

```python
plt.figure(figsize = (12,8))
sns.countplot(y = 'neighbourhood group', data=df, color='green', order=df['neighbourhood group'].value_counts().index)
plt.title('Number of listings by neighbourhood group')
plt.xlabel('count')
plt.ylabel('neighbourhood group')
plt.show()
```

# How has the number of reviews change over with time?

```python
df['last review'] = pd.to_datetime (df['last review'], errors='coerce')
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line', color='red')
plt.title('Number of reviews over time')
plt.xlabel('date')
plt.ylabel('number of reviews')
plt.show()
```
